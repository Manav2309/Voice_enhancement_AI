```python
n_fft=4096
hop_len= 512
spect=librosa.stft(audio, n_fft=n_fft, hop_length=hop_len)
spect_db=librosa.amplitude_to_db(abs(spect),ref=np.max)
```

```python
def plot_spectogram(audio):
  hop_len=512
  plt.figure(figsize=(10,4))
  librosa.display.specshow(spect_db, sr=22050, hop_length=hop_len, x_axis='time', y_axis='log')
  plt.colorbar(format='%+2.0f dB')
  plt.title('Spectrogram (dB)')
  plt.show()  
```

```python
def noise_muffle(audio):
    #adding noise
    noise_factor=0.05
    noise = np.random.normal(0, 0.1, len(audio))
    lvl1_audio=audio+noise_factor*noise
    #adding muffle
    lvl2_audio=gaussian_filter1d(lvl1_audio,sigma=2.5)    
    
    return (lvl2_audio)
```

```python
final_audio=noise_muffle(audio)
```
